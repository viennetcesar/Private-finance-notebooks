```python
import yfinance as yf
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from arch import arch_model
from statsmodels.tsa import stattools
from datetime import datetime
```

```python
ticker = 'AC.PA'

aujourdhui = datetime.today()
aujourdhui = aujourdhui.strftime("%Y-%m-%d")

data = yf.download(ticker, start='2024-01-01', end=aujourdhui)
data = data['Close']

returns = np.log(data/data.shift(1)).dropna()

plt.figure(figsize=(12, 6))
plt.plot(returns)
plt.title(f'Rendements de l\'action {ticker}')
plt.show()


plt.figure(figsize=(12,6))
plt.plot(data.index, data, color='blue')
plt.show()
```

```python
model = arch_model(returns, p=1, q=1)
result = model.fit(disp="off")

# Prévisions de volatilité
forecasts = result.forecast(horizon=10)
volatility_forecast = forecasts.variance.values

# Tracer les prévisions de volatilité
plt.figure(figsize=(12, 6))
plt.plot(volatility_forecast, label='Prévisions de Volatilité')
plt.title(f'Prévisions de Volatilité pour l\'action {ticker}')
plt.legend()
plt.show()
```
