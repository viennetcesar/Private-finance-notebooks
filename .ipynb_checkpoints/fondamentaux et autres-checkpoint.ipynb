```python
import yfinance as yf
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from arch import arch_model
from statsmodels.tsa import stattools
from datetime import datetime
```

```python
ticker = 'AC.PA'

aujourdhui = datetime.today()
aujourdhui = aujourdhui.strftime("%Y-%m-%d")

data = yf.download(ticker, start='2024-10-01', end=aujourdhui)
data = data['Close']

returns = np.log(data/data.shift(1)).dropna()

plt.figure(figsize=(12, 6))
plt.plot(returns)
plt.title(f'Rendements de l\'action {ticker}')
plt.show()


plt.figure(figsize=(12,6))
plt.plot(data.index, data, color='blue')
plt.show()
```

```python
model = arch_model(returns, p=1, q=1)
result = model.fit(disp="off")

# Prévisions de volatilité
forecasts = result.forecast(horizon=10)
volatility_forecast = forecasts.variance.values

# Tracer les prévisions de volatilité
plt.figure(figsize=(12, 6))
plt.plot(volatility_forecast, label='Prévisions de Volatilité')
plt.title(f'Prévisions de Volatilité pour l\'action {ticker}')
plt.legend()
plt.show()
```

## Mélanger GARCH et RSI et décider

```python
# Importer les bibliothèques nécessaires
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from arch import arch_model
from statsmodels.tsa.stattools import acf
from talib import RSI  # Importer TA-Lib pour calculer l'indicateur RSI

# Récupérer les données de l'action Accor depuis Yahoo Finance
ticker = 'AC.PA'  # Ticker d'Accor (CAC 40)
data = yf.download(ticker, start='2022-01-01', end='2024-01-01')

# Utiliser uniquement les prix de clôture
data = data['Close']

# Calculer les rendements
returns = np.log(data / data.shift(1)).dropna()

# Rééchelle les rendements pour éviter les problèmes de convergence
scaled_returns = returns * 100  # Multiplie les rendements par 100 pour les rééchelonner

# Estimer un modèle GARCH avec les données rééchelonnées
model = arch_model(scaled_returns, p=1, q=0)  # Ou p=0, q=1 si nécessaire
result = model.fit(disp=10)  # Augmenter disp pour obtenir plus de détails sur l'ajustement

# Prévisions de volatilité
forecasts = result.forecast(horizon=10)
volatility_forecast = forecasts.variance.values

# Calculer l'indicateur RSI
rsi = RSI(data, timeperiod=14)

# Tracer les prévisions de volatilité
plt.figure(figsize=(12, 6))
plt.plot(volatility_forecast, label='Prévisions de Volatilité', color='blue')
plt.title(f'Prévisions de Volatilité pour l\'action {ticker}')
plt.xlabel('Horizon de Prévision')
plt.ylabel('Volatilité')
plt.legend()
plt.show()

# Visualiser l'indicateur RSI
plt.figure(figsize=(12, 6))
plt.plot(data.index, data, label='Prix de Clôture', color='black')
plt.plot(data.index, rsi, label='RSI', color='blue')
plt.axhline(70, linestyle='--', color='red', label='Surachat (RSI > 70)')
plt.axhline 30, linestyle='--', color='green', label='Survente (RSI < 30)')
plt.title(f'RSI pour l\'action {ticker}')
plt.xlabel('Date')
plt.ylabel('RSI')
plt.legend()
plt.show()

# Évaluer la direction avec l'indicateur RSI
buy_signal = (rsi < 30)
sell_signal = (rsi > 70)

# Interpréter les signaux combinés
if volatility_forecast[-1] > 0.02 and sell_signal[-1]:  # Exemples de seuils pour la volatilité et RSI
    print(f'Attention, fort risque de correction à la baisse sur {ticker}.')
elif volatility_forecast[-1] > 0.02 and buy_signal[-1]:
    print(f'Opportunité d\'achat sur {ticker} avec la correction anticipée.')
else:
    print(f'Volatilité stable avec des signaux neutres sur {ticker}.')
```
