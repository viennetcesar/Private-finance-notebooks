```python
import yfinance as yf
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from arch import arch_model
from statsmodels.tsa import stattools
from datetime import datetime
```

```python
ticker = 'AC.PA'

aujourdhui = datetime.today()
aujourdhui = aujourdhui.strftime("%Y-%m-%d")

data = yf.download(ticker, start='2024-10-01', end=aujourdhui)
data = data['Close']

returns = np.log(data/data.shift(1)).dropna()

plt.figure(figsize=(12, 6))
plt.plot(returns)
plt.title(f'Rendements de l\'action {ticker}')
plt.show()


plt.figure(figsize=(12,6))
plt.plot(data.index, data, color='blue')
plt.show()
```

```python
model = arch_model(returns, p=1, q=1)
result = model.fit(disp="off")

# Prvisions de volatilit
forecasts = result.forecast(horizon=10)
volatility_forecast = forecasts.variance.values

# Tracer les prvisions de volatilit
plt.figure(figsize=(12, 6))
plt.plot(volatility_forecast, label='Prvisions de Volatilit')
plt.title(f'Prvisions de Volatilit pour l\'action {ticker}')
plt.legend()
plt.show()
```

## Mlanger GARCH et RSI et dcider

```python
# Importer les bibliothques ncessaires
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from arch import arch_model
from statsmodels.tsa.stattools import acf
from talib import RSI  # Importer TA-Lib pour calculer l'indicateur RSI

# Rcuprer les donnes de l'action Accor depuis Yahoo Finance
ticker = 'AC.PA'  # Ticker d'Accor (CAC 40)
data = yf.download(ticker, start='2022-01-01', end='2024-01-01')

# Utiliser uniquement les prix de clture
data = data['Close']

# Calculer les rendements
returns = np.log(data / data.shift(1)).dropna()

# Rchelle les rendements pour viter les problmes de convergence
scaled_returns = returns * 100  # Multiplie les rendements par 100 pour les rchelonner

# Estimer un modle GARCH avec les donnes rchelonnes
model = arch_model(scaled_returns, p=1, q=0)  # Ou p=0, q=1 si ncessaire
result = model.fit(disp=10)  # Augmenter disp pour obtenir plus de dtails sur l'ajustement

# Prvisions de volatilit
forecasts = result.forecast(horizon=10)
volatility_forecast = forecasts.variance.values

# Calculer l'indicateur RSI
rsi = RSI(data, timeperiod=14)

# Tracer les prvisions de volatilit
plt.figure(figsize=(12, 6))
plt.plot(volatility_forecast, label='Prvisions de Volatilit', color='blue')
plt.title(f'Prvisions de Volatilit pour l\'action {ticker}')
plt.xlabel('Horizon de Prvision')
plt.ylabel('Volatilit')
plt.legend()
plt.show()

# Visualiser l'indicateur RSI
plt.figure(figsize=(12, 6))
plt.plot(data.index, data, label='Prix de Clture', color='black')
plt.plot(data.index, rsi, label='RSI', color='blue')
plt.axhline(70, linestyle='--', color='red', label='Surachat (RSI > 70)')
plt.axhline 30, linestyle='--', color='green', label='Survente (RSI < 30)')
plt.title(f'RSI pour l\'action {ticker}')
plt.xlabel('Date')
plt.ylabel('RSI')
plt.legend()
plt.show()

# valuer la direction avec l'indicateur RSI
buy_signal = (rsi < 30)
sell_signal = (rsi > 70)

# Interprter les signaux combins
if volatility_forecast[-1] > 0.02 and sell_signal[-1]:  # Exemples de seuils pour la volatilit et RSI
    print(f'Attention, fort risque de correction  la baisse sur {ticker}.')
elif volatility_forecast[-1] > 0.02 and buy_signal[-1]:
    print(f'Opportunit d\'achat sur {ticker} avec la correction anticipe.')
else:
    print(f'Volatilit stable avec des signaux neutres sur {ticker}.')
```
