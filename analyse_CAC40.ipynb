```python
# On commence par créer une variable nommée cac40_moyenne_cours. Dans cette variable
# j'ai mis un {} qu'est-ce que c'est que ça ?? C'est ce qu'on appelle un dictionnaire vide en python.
# Vous vous rappelez quand je vous ai dis que [] était une liste ? Eh bien {} est un dictionnaire.
# Il est pratique d'utiliser un dictionnaire quand on veut y injecter des paires clé.valeur...
# C'est quoi une paire clé/valeur ? 
# Dans un dictionnaire Larousse version 1995 vous êtes d'accord que quand on tourne les pages
# on tombe sur un mot puis sa définition : ex : 'Salutation' : Action de saluer
# Eh bien c'est exactement pareil ici dans le code ! On utilise des dictionnaires quand on veut
# créer des couple clé (le mot) : valeur (sa définition)
# Bref pour le moment on code une ligne qui crée un dictionnaire vide et on va le remplir par la suite

cac40_moyenne_cours = {}

# Ici on réalise ce qu'on appelle une boucle en programmation. Une boucle c'est quoi ? C'est quand on
# fait plusieurs fois la même chose, on fait une chose en boucle. Par exemple on peut dire dans le code
# article par article, tu vas me lire une liste. On fait une boucle sur la liste pour lire chaque
# article. C'est hyper pratique vous verrez 

# Donc la ligne du dessous en Français : Pour chaque ticker qui se trouve dans la liste cac40_symbols_raw:
# tu vas faire ce que je te demande de faire après les ":"
for ticker in cac40_symbols_raw:
    try: # pas besoin d'expliquer le try pour le moment concntrez vous plutôt sur la suite
        
        # on crée une variable nommée data_actions dedans on lui met le cours de bourse de l'action
        # ciblée pour les 2 dernières années avec yf.Ticker(ticker).history(period='2y')
        # Au premier tour de boucle sur la liste cac40_symbols_raw la première action de la liste est
        # Accor, donc la variable en dessous va aller récupérer Accor, et son cours de bourse
        # sur les 2 dernières années
        data_action = yf.Ticker(ticker).history(period='2y') 
        
        
        # Ensuite, toujours dans cette même boucle on vérifie que le cour de bourse de Accor
        # existe bel et bien. En effet, parfois il se peut que certaines données ne soient pas disponibles
        # avec un if not empty, on verifie que le cour de bourse n'est pas "empty" -> "vide" en Français
        if not data_action.empty:
            
            # Si le cour de bourse de Accor n'est pas vide, il contient des données,
            # du coup on récupère ces données et plus précisément les données du cour de clôture
            # en sélectionnant ['Close'] pour "Clôture". Et puis comme on a toute une série de 2 ans
            # de cours de clôture pour Accor, on en fait tout simplement la moyenne avec la fonction
            # .mean() qui veut dire "moyenne" en Anglais
            moyenne_cour = data_action['Close'].mean()
            
            # Ensuite on a la moyenne mais parfois il se peut encore que le cours de bourse des 2 ans
            # pour Accor ne soit pas vide mais ait des valeurs manquantes, des jours où le cour
            # n'a pas été correctement enregistré, on créer une condition IF (SI en Français)
            # qui dit : si il n'y a pas de valeurs manquantes, alors fait ce que je te dis après les ":"
            if pd.notna(moyenne_cour):
                
                    # ENfin on donne l'instruction de récupérer notre dictionnaire vide nommé 
                    # cac40_moyenne_cours, de lui donner comme clé le ticker (AC.PA pour Accor) et
                    # lui donner une valeur après le "=" notamment la valeur de son cour moyen des
                    # 2 dernières années. Donc en résumé simple, le code cac40_moyenne_cours[ticker] = moyenne_cour
                    # permet d'ajouter à notre dictionnaire vide {Accor: 46.94€}
                    cac40_moyenne_cours[ticker] = moyenne_cour
    # Ne vous embêtez pas avec cette ligne de code
    except Exception as e:
        sys.stderr.write(f"Pour le ticker {ticker}, erreur : {e}\n")
        
# Quand le premier tour de boucle est terminé, on passe à l'action d'après dans la liste, Air liquide
# puis la troisième, la quatrième etc.... à la fin quand on a parcouru toute la liste on a un 
# sublime dictionnaire avec à gauche en clé le ticker (ex: AC.PA) et à droite en valeur le nom 
# de l'action (ex:Accor)
        
# On imprime le dictionnaire, enjoy :)        
cac40_moyenne_cours
```

### Passons aux choses sérieuses avec l'utilisation du machine learning pour tenter de prédire au plus proche l'évolution du prix d'une action. Plus spécifiquement nous allons utilise le LSTM (Long Short Term Memory) qui est un réseau de neurones et qui fait partie des méthodes de machine learning (IA) les plus récentes et précises dans la prédiction. Evidemment ce modèle n'est pas parfait et donne un résultat à la hauteur des données avec lesquelles il a été nourri. En outre, logiquement il faudrait le combiner avec d'autres modèles pour avoir des résultats plus robustes. 

##### Le code ci-dessous n'est pas commenté, il vous donne les graphiques des 40 actions du CAC40 avec leur prédiction d'évolution de cours du jour. Le point rouge est la prédiction  générée par le modèle LSTM et le trait bleu le cour réalisé. Un tel graphique nous permet de voir si notre modèle est efficace pour prédire le cours du jour ou pas. La cellule de code qui suivra celle ci-dessous est entièrement commentée et explique chaque ligne de code.
```python
# Importation des bibliothèques nécessaires
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from datetime import datetime
import plotly.graph_objects as go

# Liste des tickers CAC40
cac40_symbols_raw = [
    'AC.PA', 'AI.PA', 'AIR.PA', 'MT.AS', 'CS.PA', 'BNP.PA', 'EN.PA', 'CAP.PA', 
    'CA.PA', 'ACA.PA', 'BN.PA', 'DSY.PA', 'EDEN.PA', 'ENGI.PA', 'EL.PA', 
    'ERF.PA', 'RMS.PA', 'KER.PA', 'LR.PA', 'OR.PA', 'MC.PA', 'ML.PA', 'ORA.PA', 
    'RI.PA', 'PUB.PA', 'RNO.PA', 'SAF.PA', 'SGO.PA', 'SAN.PA', 'SU.PA', 
    'GLE.PA', 'STLA', 'STMPA.PA', 'TEP.PA', 'HO.PA', 'TTE.PA', 'URW.PA', 
    'VIE.PA', 'DG.PA', 'VIV.PA'
]

# Paramètres généraux
start_date = datetime(datetime.now().year, 1, 1)
end_date = datetime.now()

# Partie 1 : Téléchargement des données historiques pour toutes les actions
print("Téléchargement des données...")
data_dict = {}
for ticker in cac40_symbols_raw:
    data = yf.download(ticker, start=start_date, end=end_date)['Close']
    data_dict[ticker] = data

# Fusionner toutes les séries de clôture dans un seul DataFrame
combined_data = pd.DataFrame(data_dict)

# Partie 2 : Visualisation des cours de clôture pour chaque action
plt.figure(figsize=(14, 8))
for ticker in cac40_symbols_raw:
    plt.plot(combined_data.index, combined_data[ticker], label=ticker, linewidth=1)
plt.title("Cours de clôture des actions du CAC40 en " + str(datetime.now().year))
plt.xlabel("Date")
plt.ylabel("Prix de clôture (€)")
plt.legend(fontsize='xx-small', loc="upper left")
plt.grid()
plt.show()

# Partie 3 : Prédiction avec le modèle LSTM
def create_lstm_model():
    model = Sequential()
    model.add(LSTM(units=50, return_sequences=True, input_shape=(60, 1)))
    model.add(LSTM(units=50, return_sequences=False))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

# Prédiction pour chaque action
for ticker in cac40_symbols_raw:
    print(f"\nEntraînement du modèle LSTM pour {ticker}...")
    
    # Préparation des données
    data = combined_data[ticker].dropna()
    dataset = data.values.reshape(-1, 1)
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(dataset)
    
    # Création des séquences
    prediction_days = 60
    x_train, y_train = [], []
    for i in range(prediction_days, len(scaled_data)):
        x_train.append(scaled_data[i-prediction_days:i, 0])
        y_train.append(scaled_data[i, 0])
    
    x_train, y_train = np.array(x_train), np.array(y_train)
    x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))
    
    # Entraînement du modèle
    model = create_lstm_model()
    model.fit(x_train, y_train, epochs=5, batch_size=32, verbose=0)
    
    # Prédiction
    test_data = scaled_data[-prediction_days:]
    x_test = np.array([test_data])
    x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))
    
    predicted_price = model.predict(x_test)
    predicted_price = scaler.inverse_transform(predicted_price)
    
    # Affichage des résultats
    plt.figure(figsize=(10, 6))
    plt.plot(data.index[-200:], data[-200:], color='blue', label='Prix réel')
    plt.axvline(data.index[-1], color='gray', linestyle='--', label="Aujourd'hui")
    plt.scatter(data.index[-1], predicted_price[0, 0], color='red', label='Prédiction LSTM')
    plt.title(f"Prédiction LSTM pour {ticker}")
    plt.xlabel("Date")
    plt.ylabel("Prix de clôture (€)")
    plt.legend()
    plt.grid()
    plt.show()

print("Analyse et prédiction terminées pour toutes les actions.")
```

On remarque que le modèle se plante souvent, certains jours il prédit une forte hausse du cours alors que c'est une baisse qui se réalise.... Pourquoi ? 

### Pourquoi les modèles LSTM sont moins efficaces pour prédire le prix absolu d'une action :

1. **Nature des Données** :
   - **Prix absolu d'une action** : Le prix d'une action est généralement influencé par des facteurs multiples, y compris les événements économiques, les rapports financiers, les résultats de l'entreprise, les annonces de bénéfices, etc. Ces influences peuvent être complexes et difficiles à capturer avec des modèles basés uniquement sur les données historiques de prix.
   - **Effet d'entraînement** : Les prix absolus peuvent fluctuer considérablement et peuvent être affectés par de nombreux bruits et irrégularités dans les données historiques. Les LSTM, bien que capables de capturer des dépendances temporelles, peuvent mal généraliser sur des séries chronologiques de prix spécifiques en raison de ces fluctuations.

2. **Caractérisation des séries temporelles** :
   - Les prix absolus peuvent ne pas suivre une tendance claire ou peuvent être influencés par des variations à court terme difficilement modélisables. Par exemple, un pic de prix à cause d'une annonce spécifique ou d'un événement imprévu peut perturber les modèles LSTM qui cherchent à identifier des modèles de comportement plus généraux et stables.

### Pourquoi les modèles LSTM peuvent être efficaces pour prédire le rendement :

1. **Prévision de Rendement** :
   - **Rendement d'une action** : Il se réfère généralement à la variation du prix par rapport à un certain intervalle de temps. Les modèles LSTM peuvent mieux capturer des séquences de variations relatives des prix et les relations temporelles sous-jacentes entre les rendements.
   - **Stabilité des rendements** : Les rendements sont souvent moins volatils que les prix absolus, ce qui peut permettre aux LSTM de mieux généraliser et de faire des prédictions plus précises.

2. **Dépendances temporelles** :
   - Les LSTM sont conçus pour apprendre des dépendances temporelles complexes dans les données. Cette capacité les rend particulièrement adaptés à la prédiction de séquences comme les rendements, où il est important de considérer l'historique pour anticiper les futures variations.

#### Donc un prix est difficile à prédire car si vous regarder les cours d'une action il est possible que sur certaines périodes ce soit presque le calme plat puis un jour soudainement une explosion à la hausse. Si on a entraîné le modèle LSTM sur une fenêtre pendant laquelle c'était calme, alors le modèle ne pourra pas prédire une explosion à la hausse, il sera surpris. La différence avec un entrainement sur des rendements plutôt que sur le cours lui même c'est que les rendements sont en général bornés à la hausse et à la baisse. Ils sont stationnaires. En outre, parfois plusieurs hausse successives se produisent, ou plusieurs baisses succéssives. Le modèle LSTM peut logiquement mieux capter certaines tendances et donc mieux performer en prédiction. 

Intégrer d'autres paramètres que le seul prix dans LSTM

```python
# On importe tous les modules de code qui vont nous simplifier la tâche dans notre analyse
# matplotlib par exemple nous permettra de tracer des graphiques sans trop forcer et tensorflow
# réalisera le mondèle complet LSTM sans que nous ayons à le programmer nous même à la main.
# Il suffira de donner les bons paramètres au modèle et il calculera tout par lui même.
# yfinance nous permet de récupérer des données de bourse sur les entreprises souhaitées

import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
import matplotlib.dates as mdates


# Concept un peu particulier ici, on initialise un état pour que notre modèle commence toujours
# à partir du même état et donne toujours la même prédiction pour le même jour
np.random.seed(42)  

# On teste notre modèle avec les données financières de 
ticker = "AC.PA"
start_date = "2023-10-01" # choisit la date à partir de laquelle on veut récupérer les données 

# On utilise le module pd pour lui demander de récupérer la date actuelle pour lancer notre modèle
# pourquoi ? Parce que si vous décidez de lancer votre code dans 1 an alors il serait cool que la
# date soit actualisée afin que vous récupériez directement les bonnes données
# avec pd.Timestamp.now() on récupèr le temps maintenant (now), ce qui veut dire qu'on récupère
# la date d'aujourd'hui. 
# Ensuite le strftime permet de donner à la date une forme particulière très précise que nous choisissons
# nous même. Ici on choisit nous même le format de la date en disant entre parenthèses qu'on veut 
# l'année-mois-jour. Le %Y signifie "Year" (année) le %m signifie "month" le mois et le %d signifie
# "day" le jour. 
end_date = pd.Timestamp.now().strftime('%Y-%m-%d')  

# avec le module yf on peut télécharger (download) les données financière. On doit juste préciser entre
# paranthèses les données de quelle action en donnant son "ticker", la date de début de la série
# qu'on veut récupérer et la date de fin. 
# On capture cette série de données et on l'encapsule dans une varibale qu'on appelle stock_data
stock_data = yf.download(ticker, start=start_date, end=end_date)

# Calcul des rendements car on va tenter de prédire les rendements
# Pour calculer un rendement il faut tout simplement calculer une variation. C'est à dire:
# si l'action vaut 110 aujourd'hui et qu'elle valait 100 hier alors la variation est de (110-100)/100
# donc 10/100 donc 10%
# Et comme on expliquait dans le texte au dessus de cette cellule de code une variation se prédit déjà
# mieux que le cours d'une action. Il peut donc être très intéressant de prédire le rendement et de
# tenter d'anticiper par exemple que demain l'action devrait augmenter de 2%
# Dans le code ci dessous on récupère la série de données capturée précedemment dans la variable 
# stock_data et plus particulièrement on récupère dans cette série le cours de clôture en sélectionnant
# ['Close'] pour clôture. En effet, dans la variable stock_data il n'y a pas que le cours de clotûre
# il y a le cours haut le cours bas etc...
# Ensuite on utilise la fonction .pct_change() qui calcule justement le rendement, c'est à dire de
# combien l'action a varié d'un jour à l'autre. 
stock_data['Return'] = stock_data['Close'].pct_change()

# Si il manque certaines données de cour de bourse dans le jeu de données ça va être compliqué
# de calculer les rendements. Par exemple si l'action vaut 100 le 2 janvier et que le 3 janvier il 
# n'y a pas de donnée, il est impossible de faire un calcul de rendement. Donc par défaut
# python mettra dans la cellule qui n'a pas de données la valeu "NaN" qui signifie Not A Number 
# (pas un nombre)
# avec la ligne de code ci-dessous on élimine toutes les cellules dans lesquelles on a NaN grâce
# à la fonction .dropna()
# Enfin, ces cellules étant éliminées les données sont complètes et on peut calculer des rendements
stock_data = stock_data.dropna()  



# Création des séquences pour le modèle LSTM
def create_sequences(data, time_steps=60):
    
    # on commence par préparer 2 listes vides qu'on va remplir, une pour les X et une pour les Y
    X, y = [], []
    
    
    for i in range(len(data) - time_steps - 1):  
        
        # rendement des jours 1 à 60 qui permettent d'entraîner le modèle
        X.append(data[i:i + time_steps, 0])
        
        # rendement du jour 61 pour que le modèle prédise le jour 61 à partie des 60 précédents
        y.append(data[i + time_steps + 1, 0])  
        
        # à la fin la fonction nous retourne nos données sous forme de tableau ARRAY
        # elle retourne un tableau pour les valeurs X et un tableau pour les valeurs Y
    return np.array(X), np.array(y)

time_steps = 60  # Longueur de la séquence pour le LSTM, pour son apprentissage

# in utilise la fonction crée ci-dessus create_sequences qui permettra au modèle de s'entraîner
X, y = create_sequences(stock_data[['Return']].values, time_steps)

# Code un peu particulier qui permet de donner le bon format à notre X pour que le modèle s'entraîne
# Reshape pour le LSTM : (samples, time_steps, features)
X = X.reshape(X.shape[0], X.shape[1], 1)

# Division des données : entraînement (80%) et test (20%)
# On divise les données en conservant l'ordre. En effet c'est une série temporelle l'ordre est très
# important
# pourquoi on divise les données ? Parce que la convention est d'entrainer le modèle sur une partie
# seulement des données (80% ici) et tester le modèle sur les 20% de données restantes, c'est à dire 
# des données que le modèle n'a jamais vues et dont nous connaissons à l'avance l'issue.
# En effet, tester le modèle sur les 20% de données qu'il ne connait et que nous connaissons nous
# permet de vérifier à quel point il est bon ou mauvais. 

# ici on récupère la longueur de notre jeu de données avec le fonction len(X) len pour length en anglais
# qui veut dire longueur, puis on prend seulement 0.8 de cette longueur, autrement dit 80%
# enfin, il se peut que ce chiffre 0.8 x len(X) ne soit pas un nombre entier, par exemple si le jeu
# de données a une longueur de 122 cours de bourse , 80% de 122 = 97,6. Du coup avec la fonction
# int() on prend sa valeur entière, 97 ou 98 pour pouvoir prendre un nombre de lignes entiers dans
# le jeu de données. 
train_size = int(len(X) * 0.8)

# Une fois que dans la variable train_size on a un chiffre égal à 80% de la taille du jeu de données, 
# on crée une variable X qui contient 80% du jeu de données et une variable X qui contient le reste
# par exemple : si on a 120 jours de données qui vont du 1er janvier 2024 au 30 avril 2024 alors dans 
# le 1er X on aura 80% des 120 jours c'est à dire les 96 premiers jours [1janvier-6avril] et dans le 
# le 2ème X, le X_test, on aura les 20% de jours restants c'est à dire du 6avril au 30avril.
# Le modèle s'entraînera uniquement sur X_train donc les données du 1er janvier au 6avril.
# Plus concrètement, sans parler date directement mais plutôt en parlant ""numéro de ligne" dans le
# jeu de données, dans X_train on récupère les données de la ligne 1 à la ligne 96, et dans X_test
# les donnnées de la ligne 96 à la ligne 120.
# Si on reprend notre exemple et que la taille du jeu de données est de 120 lignes (1janvier au 30avril)
# alors la variables train_size qui sélectionne 80% des lignes contiendra 96 lignes et X_test 24 lignes
# donc concrètement "train_size=96"
# et pour sélection les 96 premières lignes de X on fait X[:train_size] les ":" veulent dire "prends 
# moi toutes les lignes jusqu'à 96 ". et X[train_size:] veut dire prends moi toutes les lignes de
# la ligne 96 jusqu'à la fin
X_train, X_test = X[:train_size], X[train_size:]

# Même raisonnement pour les Y, ici on ne prend pas les dates mais les valeurs, c'est à dire le cour 
# de bourse. Donc finalement dans le X on a la date et dans le Y le cour de bourse. C'est bon on est
# prêt pour entrainer le modèle LSTM !!!!!!
y_train, y_test = y[:train_size], y[train_size:]


# Partie complexe du modèle IA, ici on créer notre modèle LSTM Long Short Term Memory
# qui est un réseau de neurones avec mémoire courte et mémoire longue. On ne va pas rentrer dans les
# détails mais en version simple le réseau de neurones va apprendre des données et ajuster le poids
# dédié aux neurones par une minimisation des erreurs. Le modèle particulier LSTM comme son nom
# l'indique va capter une tendance LONG TERME et une tendance COURT TERME et il va tenter à chaque
# nouvelle avancée dans les données de bourse de comprendre si la nouvelle donnée de bourse (le cour)
# est due à un passé récent (SHORT TERM) ou à un passé lointain (LONG TERM) et le modèle va donner
# des poids au passé lointain et au passé proche. Ce modèle est puissant dans le sens où il s'adapte
# vraiment aux données (Machine Learning). C'est à dire qu'il est capable de capter si le passé lointain
# est en effet responsable pour une partie des mouvement du cour de bourse ou pas. Par exemple si 
# l'action valait 100 hier et qu'aujourd'hui elle chute à 20 comme dans le cas de la réccente scission
# de VIVENDI, le modèle comprend que ce n'est pas le passé qui est responsable de cette chute mais
# une nouvelle récente et choquant, il intègre donc un fort poids dans la passé proche et diminue
# le poids associé au passé lointain. Il apprend de manière intelligente par une approche qui utilise
# toutes les données et accorde des poids c'est bon pour l'idée ?

# On commence par initialiser une succession de couches vides avec Sequential()
model = Sequential()

# On ajoute une première couche avec 50 neurones, on lui demande return_sequences=True qu'il traite 
# chaque entrée sans exception (chaque cour de bourse donné) et input_shape=(time_steps, 1) précise
# qu'on donné à notre modèle des données qui on la forme suivant : time_steps = 60 cours de bourse
# et  , 1 signifie qu'on donne à notre modèle 1 seule variable (le cour de bourse) bien qu'il y en ait
# 60 à la suite on ne lui donne qu'1 variable cours de bourse. On aurait trsè bien pu lui donner
# un cours de bourse + le nombre d'actions achetées + la moyenne mobile, bref, d'autres variables
model.add(LSTM(50, return_sequences=True, input_shape=(time_steps, 1)))

# 50 unités : Cette couche LSTM a également 50 neurones.
# return_sequences=False : Cela signifie que la couche renvoie une seule sortie (un vecteur final) 
# pour toute la séquence d'entrée. Elle ne produit pas une séquence comme la première couche.
model.add(LSTM(50, return_sequences=False))

# 25 neurones : Cette couche dense (fully connected) contient 25 neurones.
# Entrée : Elle reçoit le vecteur de sortie de la couche LSTM précédente (taille (50)).
# Fonctionnement : Chaque neurone effectue un calcul pondéré de l'entrée et applique une activation 
# (par défaut, ReLU si non spécifié).
# Sortie : La sortie est un vecteur de taille (25).
# La couche dense est celle qui contient la fonction d'activtion Relu dans le réseau de neurones
# celle qui permet de traiter le NON LINEARITE des données de bourse
model.add(Dense(25))

# 1 neurone : Cette couche produit une seule valeur comme sortie, utile pour des tâches de régression 
# ou de classification binaire.
# Entrée : Elle reçoit le vecteur de sortie de taille (25) de la couche précédente.
# Sortie : Un scalaire (valeur unique).
# On ajoute cette 2ème couche dense pour mieux comprendre les données
model.add(Dense(1))



# Maintnenant qu'on a préparé le nombre de couches du modèles on va lui donner des conditions
# en explicitant la manière dont on veut le faire travailler pour qu'il s'entraîne selon nos
# désirs. Ici encore le vocubulaire devrait vous sembler obscure car il nécessite d'être initié
# au machine learning. Pas grave, comprendre l'idée générale c'est déjà cool. En gros, au delà
# de donner à notre modèle nombre de couches de neurones pour qu'il apprenne, on va aussi lui donner
# des règles à suivre. Par exemple on peut lui dire de quelle manière il devra calculer les poids
# à affecter aux neurones. En effet, il existe plusieurs manières qui ont leurs avantages et inconvénients

# compile configure le modèle pour l'entraînement.
# optimizer='adam' :
# Adam est un algorithme d'optimisation très utilisé pour ajuster les poids du réseau.
# Il combine les avantages de deux autres algorithmes : RMSprop et SGD avec momentum.
# Adam ajuste le taux d'apprentissage automatiquement, ce qui le rend efficace pour divers problèmes.
# loss='mean_squared_error' :
# La fonction de perte choisie ici est l'erreur quadratique moyenne (MSE).
# MSE mesure l'écart moyen entre les prédictions du modèle et les vraies valeurs cibles.
# Idéal pour les problèmes de régression.
model.compile(optimizer='adam', loss='mean_squared_error')


# fit entraîne le modèle sur les données d'entrée 𝑋 train et leurs sorties associées 𝑦 train
# Paramètres principaux :
# batch_size=16 :
# Pendant l'entraînement, les données sont divisées en petits groupes appelés batchs.
# Ici, le modèle met à jour ses poids après avoir vu 16 échantillons.
# epochs=10 :
# Le modèle verra les mêmes données 10 fois pendant l'entraînement.
# Chaque "passage" complet sur les données est appelé une époque.
# validation_data=(X_test, y_test) :
# Ces données ne participent pas à l'entraînement mais sont utilisées pour évaluer la 
# performance du modèle après chaque époque.
# Résultat :
# history contient des informations sur la progression de l'entraînement :
# La valeur de la fonction de perte(MSE) pour les données d'entraînement et de validation à chaque époque.
history = model.fit(X_train, y_train, batch_size=16, epochs=10, validation_data=(X_test, y_test))


# predict génère des prédictions pour les entrées 𝑋 test Chaque échantillon de 𝑋 test donne une 
# sortie (par exemple, une valeur de retour ou un prix prédit).
# Rappelez vous que le modèle a apprise avec les données X_train et ne connait pas les données
# X_test c'est donc ici qu'on teste le modèle en lui donnant les données X_test et en lui demandant
# de faire une prédiction du cour de bourse (ou rendement) à partir des dates contenues dans X_test
predictions = model.predict(X_test)

# On commence à prédire à partir de train_size + time_step car on a besoin de la zone time_step
# pour apprendre et on prédit timestep + 1 bref, restons cool vous aurez tout le temps d'apprendre
# avec des tutos
predictions = stock_data['Return'].iloc[train_size + time_steps + 1:] 

# pour faire les prédiction du futur il faut forcément récupérer les dernier time_step ici 60jours
# étant donné que le modèle s'est entrainé en utilisant toujours les 60 derniers jours il faut le 
# nourrir des 60 derniers jours pour prédire les suivants
last_known_returns = stock_data['Return'].iloc[-time_steps:].values

# ici on prépare une variable qui contient combien de jours on veut prédire (2), avec quelle fenêtre
# de données passées (time_step=60) et en lui précisant où on commence à prédire (last_known_return)
# np.array sert à mettre le données en forme de tableau
X_future = np.array([last_known_returns] * 2).reshape(2, time_steps, 1)


# On utilise ces données des 60 derniers jours pour réaliser la prédiction finale du rendement
future_predictions = model.predict(X_future)

# les rendements se cumulent dans la vraie vie donc on a similé des rendements jour après jour
# ici on les cumule pour vraiment obtenir notre rendement final cumulé
future_returns = stock_data['Return'].iloc[-1] + np.cumsum(future_predictions.flatten())


# On génère les dates des 2 prochains jours pour pouvoir dessiner sur un graphique ensuite
# le cour de bourse et sa date
future_dates = pd.date_range(stock_data.index[-1] + pd.Timedelta(days=1), periods=2)

# Pour visualiser les résultat on prépare un graphique et on lui précise la taille du graphique
# à l'aide du module plt et de la fonction .figure. Entre parathèses on donne la taille explicite
# de la figure, ici 12 sur 6
plt.figure(figsize=(12, 6))

# On dessine sur le graphique le rendement (capturé précédemment dans la variable stock_data['Return'])
# et on précise qu'on veut dessiner le série en couleur "bleue" avec color='blue'
# puis on précise que notre série de bourse bleue s'appellera 'Rendement Historique' 
plt.plot(stock_data.index, stock_data['Return'], color='blue', linewidth=2, label='Rendement Historique')

# On dessine ici les prédictions réalisées par notre modèle LSTM en rouge. 
# On nomme la série rouge explicitement 'Prédiction LSTM'
plt.plot(stock_data.index[train_size + time_steps + 1:], predictions, color='red',
         linewidth=0.5 ,label='Prédiction LSTM')

# Prédictions futures de rendements
plt.plot(future_dates, future_returns, color='purple', label='Prédiction 2 jours')

# Formatage des dates sur l'axe x, avec le module plt on utilise la fonction préexistante xticks
# qui permet de donner des directives à l'axe horizontal du graphique. Ici on lui
# demande de tourner les dates à 45° pour le style
plt.xticks(rotation=45)
         
# toujours avec le module plt et à l'aide de la fonction gca() on s'attaque encore à l'axe horizontal 
# xaxis et on lui donne le format explicite de la date telle qu'on souhaite la voir apparaître
# %d/%m/%y pour day (jour), month (mois) et year (année)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d/%m/%y'))

# On ajoute un titre sur le graphique avec .title
# puis on ajoute un titre à l'axe des x (horizontal) et un titre à l'axe des y (vertical)
# On ajoute une belle légende et une grille tracée en fond. Enfin on demande à python de tracer le 
# graphique explicitement avec plt.show()
plt.title(f"Prédiction des rendements pour {ticker} - 2 jours futurs")
plt.xlabel("Date")
plt.ylabel("Rendement")
plt.legend()
plt.grid()
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM
import matplotlib.dates as mdates


ticker = "AC.PA"
start_date = "2023-10-01" 

end_date = pd.Timestamp.now().strftime('%Y-%m-%d')  
stock_data = yf.download(ticker, start=start_date, end=end_date)
stock_data['Return'] = stock_data['Close'].pct_change()
stock_data = stock_data.dropna()  


def create_sequences(data, time_steps=60):    
    X, y = [], []    
    for i in range(len(data) - time_steps - 1):    
        X.append(data[i:i + time_steps, 0])
        y.append(data[i + time_steps + 1, 0])          
    return np.array(X), np.array(y)


time_steps = 60  

X, y = create_sequences(stock_data[['Return']].values, time_steps)
X = X.reshape(X.shape[0], X.shape[1], 1)
train_size = int(len(X) * 0.8)
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

model = Sequential()
model.add(LSTM(50, return_sequences=True, input_shape=(time_steps, 1)))
model.add(LSTM(50, return_sequences=False))
model.add(Dense(25))

model.add(Dense(1))
model.compile(optimizer='adam', loss='mean_squared_error')
history = model.fit(X_train, y_train, batch_size=16, epochs=10, validation_data=(X_test, y_test))
predictions = model.predict(X_test)
predictions = stock_data['Return'].iloc[train_size + time_steps + 1:] 


last_known_returns = stock_data['Return'].iloc[-time_steps:].values
X_future = np.array([last_known_returns] * 2).reshape(2, time_steps, 1)
future_predictions = model.predict(X_future)
future_returns = stock_data['Return'].iloc[-1] + np.cumsum(future_predictions.flatten())
future_dates = pd.date_range(stock_data.index[-1] + pd.Timedelta(days=1), periods=2)


plt.figure(figsize=(12, 6))
plt.plot(stock_data.index, stock_data['Return'], color='blue', linewidth=2, label='Rendement Historique')
plt.plot(stock_data.index[train_size + time_steps + 1:], predictions, color='red',
         linewidth=0.5 ,label='Prédiction LSTM')
plt.plot(future_dates, future_returns, color='purple', label='Prédiction 2 jours')
plt.xticks(rotation=45)
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%d/%m/%y'))
plt.title(f"Prédiction des rendements pour {ticker} - 2 jours futurs")
plt.xlabel("Date")
plt.ylabel("Rendement")
plt.legend()
plt.grid()
plt.show()
```
